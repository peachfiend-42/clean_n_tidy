```python
# fire up the pandas
import pandas as pd
```

```python
# get csv data
hotel_bookings = pd.read_csv('hotel_bookings.csv')
```

```python
# peek at data
hotel_bookings
```

```python
# drop and/or rename column
hotel_bookings.columns
```

```python
hotel_bookings.rename({'adults': 'num_adults', 'children': 'num_children', 'babies': 'num_babies'}, axis=1, inplace=True)
```

```python
# null values
hotel_bookings.isnull().sum() *100 / len(hotel_bookings)
```

```python
hotel_bookings['agent'].unique()
hotel_bookings[hotel_bookings['agent'] == 5]

# hotel_bookings['agent'].fillna(-1, inplace=True)
hotel_bookings.fillna({'agent': -1}, inplace=True)

hotel_bookings[hotel_bookings['num_children'].isna()]

hotel_bookings[hotel_bookings['country'].isna()][['hotel', 'is_canceled', 'country']]

hotel_bookings.dropna(subset=['num_children'], inplace=True)

hotel_bookings.dropna(subset=['country'], inplace=True)

```

```python
hotel_bookings.drop(columns='company', inplace=True)
```

```python
# check column data
hotel_bookings.dtypes
```

```python
hotel_bookings = hotel_bookings.astype({'is_canceled': 'boolean', 'is_repeated_guest': 'boolean', 'num_children': 'int'})
```

```python
# group columns
hotel_bookings['lead_time'].unique()
hotel_bookings['lead_time'].describe()

bins = [0, 100, 200, 300, 400, 500, 600, 700, 800]
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800']
hotel_bookings['lead_time_tiers'] = pd.cut(hotel_bookings['lead_time'],bins=bins,labels=labels)

hotel_bookings[['lead_time', 'lead_time_tiers']]
```

```python
hotel_bookings['hotel'] = hotel_bookings['hotel'].replace(r"[\*\n\^]",'',regex=True)
```
